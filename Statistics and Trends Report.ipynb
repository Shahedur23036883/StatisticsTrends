## Introduction

The dataset we have selected for our report is, Temperature and Ice Cream Sales (manayon, 2024). In this dataframe there are only two variables, Temperature and Ice Cream Profits.

Ice cream sales are often influenced by various variables, with temperature being a significant variable. As temperatures rise, people try to cool themselves with ice cream. Which makes ice cream a popular choice during warmer seasons. Understanding the relationship between temperature and ice cream profits is essential for businesses in the ice cream industry to optimize their strategies and maximize the profits.

In this report, we analyze a dataset that records ice cream profits alongside corresponding temperatures. By examining this data, we aim to uncover trends and try to get insights using various plots that can help businesses make informed decisions regarding pricing, marketing campaigns, and inventory management. Through statistical analysis and visualization techniques. we explore the impact of temperature fluctuations on ice cream sales and identify potential opportunities for enhancing profit generation.

```python
# Import necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the csv file using pandas library

temp_ice = pd.read_csv("Ice Cream Sales - temperatures.csv")

temp_ice.head()
```

```python
def temp_ice_line_graph(x, y):
    """Create a line plot with ice cream profits in the x axis and temperature in the y axis."""
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, marker='o', linestyle='-')
    plt.xlabel('Ice Cream Profits')
    plt.ylabel('Temperature')
    plt.title('Ice Cream sales goes up when Temperature is high')
    plt.grid(True)
    plt.show()
    return
```

```python
temp_ice_line_graph(temp_ice['Ice Cream Profits'], temp_ice['Temperature'])
```

```python
def temp_ice_bar_plot(x, y):
    """Create a bar plot of Ice cream sales vs temperature."""
    plt.figure(figsize=(8, 6))
    plt.bar(x, y)
    plt.xlabel('Temperature')
    plt.ylabel('Ice Cream Profits')
    plt.title('Ice Cream Sales goes up when Temperature is High')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()
    return
```

```python
temp_ice_bar_plot(temp_ice['Temperature'], temp_ice['Ice Cream Profits'])
```

The line plot and the bar plot indicate that as the temperature increases, ice cream sales also increase. The highest sales occurs at 98 degrees. This suggests that ice cream sales are positively correlated with temperature, meaning that hotter temperatures lead to higher ice cream sales. The profits seem to follow the same trend as the sales, with the highest profits.

Therefore, the conclusion we can make from this bar plot is that ice cream sales and profits are higher when the temperature is also higher. This information could be useful for ice cream vendors to optimize their sales and inventory management.

```python
def temp_ice_violin_plot(data):
    """Create a violin plot."""
    plt.figure(figsize=(8, 6))
    plt.violinplot(data, showmeans=True)
    plt.xlabel('Temperature ')
    plt.ylabel('Ice Cream Sales')
    plt.title('Temperature vs Ice Cream Sales Violinplot')
    plt.grid(axis='y')
    plt.show()
    return
```

```python
temp_ice_violin_plot(temp_ice)
```
